```r
RTS_TRAIN = na.omit(RTS_TRAIN)
```

```r
RTS_TEST = na.omit(RTS_TEST)
```

```r
str(RTS_TRAIN)
```

```r
str(RTS_TEST)
```

```r
RTS_TRAIN
```

```r
# #Setting the reference variables
RTS_TRAIN$raceeth = relevel(RTS_TRAIN$raceeth, "White")
RTS_TEST$raceeth = relevel(RTS_TEST$raceeth, "White")
```

```r
# #Linear Regression
lmScore = lm(readingScore ~ ., data = RTS_TRAIN)
```

```r
summary(lmScore)
```

```r
lmScore$residuals
```

```r
sqrt(mean(lmScore$residuals^2))
```

```r
# #ANS : 73.36555
```

```r
29.542707*2
```

```r
# #ANS : 59.085414
```

```r
predTest = predict(lmScore, newdata = RTS_TEST)
```

```r
summary(predTest)
```

```r
637.7 - 353.2
```

```r
sum((predTest - RTS_TEST$readingScore)^2)
```

```r
sqrt(mean((predTest - RTS_TEST$readingScore)^2))
```
