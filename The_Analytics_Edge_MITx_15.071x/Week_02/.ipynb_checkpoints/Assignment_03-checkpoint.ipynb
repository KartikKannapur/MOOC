```r
# #log(ILI) ~ Intercept + Coeff * Queries
```

```r
# #Linear Regression
FluTrend1 = lm(log(ILI) ~ Queries, data = FluTrain)
```

```r
summary(FluTrend1)
```

```r
cor_var = cor(log(FluTrain$ILI), FluTrain$Queries)
```

```r
cor_var
```

```r
cor_var^2
```

```r
FluTest = read.csv("FluTest.csv")
```

```r
PredTest1 = exp(predict(FluTrend1, newdata = FluTest))
```

```r
summary(PredTest1)
```
