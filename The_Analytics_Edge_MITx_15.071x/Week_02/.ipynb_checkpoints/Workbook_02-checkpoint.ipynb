# Correlation and Multicollinearity

```r
wine = read.csv("wine.csv")
```

```r
# #Correlation
?cor
```

```r
# #Correlation
cor(wine$WinterRain, wine$Price)
```

```r
cor(wine$Age, wine$FrancePop)
```

```r
# #Correlation between all variables
cor(wine)
```

# MAKING PREDICTIONS

```r
model_6 = lm(wine$Price ~ (wine$AGST + wine$HarvestRain + wine$WinterRain + wine$Age), data = wine)
```

```r
summary(model_6)
```

```r
wine_test = read.csv("wine_test.csv")
```

```r
str(wine_test)
```

```r
predict(model_6, newdata = wine_test)
```
