```r
# #Splitting Data into Training and Testing Sets
```

```r
SongsTrain = subset(songs, year < 2010)
```

```r
nrow(SongsTrain)
```

```r
SongsTest = subset(songs, year == 2010)
```

```r
nrow(SongsTest)
```

```r
# #Logistic Regression
```

```r
non_variables = c("year", "songtitle","artistname", "songID", "artistID")
```

```r
non_variables
```

```r
# #Removing non_variables from the dataset
SongsTrain = SongsTrain[ , !(names(SongsTrain) %in% non_variables) ]
SongsTest = SongsTest[ , !(names(SongsTest) %in% non_variables) ]
```

```r
# #We will only use those values in our dataset that describe numerical values
model_1 = glm(Top10 ~ ., data = SongsTrain, family = "binomial")
```

```r
summary(model_1)
```

```r
cor(SongsTrain)
```

```r
# #Taking care of Multicollinearilty
SongsTrain_2 = SongsTrain[ , !(names(SongsTrain) %in% c("loudness")) ]
SongsTest_2 = SongsTest[ , !(names(SongsTest) %in% c("loudness")) ]
```

```r
SongsTrain_3 = SongsTrain[ , !(names(SongsTrain) %in% c("energy")) ]
SongsTest_3 = SongsTest[ , !(names(SongsTest) %in% c("energy")) ]
```

```r
model_2 = glm(Top10 ~ ., data = SongsTrain_2, family = "binomial")
```

```r
model_3 = glm(Top10 ~ ., data = SongsTrain_3, family = "binomial")
```

```r
summary(model_2)
```

```r
summary(model_3)
```

```r
# #PREDICTION
```

```r
prediction_3 = predict(model_3, newdata = SongsTest,type = "response")
```

```r
prediction_3
```

```r
table(SongsTest$Top10, prediction_3 >=0.45)
```
