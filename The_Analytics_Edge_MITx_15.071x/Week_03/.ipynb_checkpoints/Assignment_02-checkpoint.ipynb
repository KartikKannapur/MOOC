# PREDICTING PAROLE VIOLATORS

```r
parole = read.csv("parole.csv")
```

```r
str(parole)
```

```r
table(parole$violator)
```

```r
summary(parole)
```

```r
parole
```

```r
# #There are two unordered variables - Crime and State
```

```r
# #Converting these two variables into Factors
parole$crime = as.factor(parole$crime)
parole$state = as.factor(parole$state)
```

```r
summary(parole)
```

```r
# #Splitting the data into Training and Testing Sets
```

```r
# #Different values of seed, will lead to different outcomes
set.seed(144)
```

```r
library(caTools)
```

```r
split = sample.split(parole$violator, SplitRatio = 0.7)
```

```r
train = subset(parole, split == TRUE)
test = subset(parole, split == FALSE)
```

```r
nrow(train)
```

```r
nrow(test)
```

```r
473/(473+202)
```

```r
# #Logistic Regression
```

```r
model_1 = glm(violator ~ ., data = train, family = "binomial")
```

```r
summary(model_1)
```

```r
log(1.61)
```

```r
# #Consider a parolee who is male, of white race, aged 50 years at prison release, from the state of Maryland, served 3 months,
# #had a maximum sentence of 12 months, did not commit multiple offenses, and committed a larceny. 
```

```r
log_equn = -4.2411574 + (0.3869904*1) + (0.8867192*1) + (-0.0001756*50) + (-0.1238867*3) + (0.0802954*12) + (0.6837143)
```

```r
log_equn
```

```r
odds = exp(log_equn)
```

```r
odds
```

```r
prob = 1/(1+ exp(-1*log_equn))
```

```r
prob
```

```r
# #TESTING
```

```r
predictParole = predict(model_1, newdata = test, type = "response")
```

```r
max(predictParole)
```

```r
table(test$violator, predictParole >= 0.5)
```
